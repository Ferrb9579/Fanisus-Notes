Missing data analysis

```python
import pandas as pd
df=pd.read_csv('Toyota.csv',na_values=["??","????"],index_col=[0])
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df[df['Age'].isnull()].head(10)
```

Drop missing values

```python
#Drop rows with null values
df1=df.copy()
display(df1.shape)
#syntax
df1.dropna(axis=0,inplace=True)
display(df1.shape)
```

```python
#Drop columns with null values
df1=df.copy()
display(df1.shape)
#syntax
df1.dropna(axis=1,inplace=True)
display(df1.shape)
```

```python
#Drop any rows that contains an NaN value for the 'MetColor' column 
df1=df.copy()
display(df1.shape)
#syntax
df1.dropna(how='any',subset=['MetColor'],inplace=True)
display(df1.shape)
```

Impute the missing values

```python
#fill all NaN value with 'Empty'
import numpy as np
df1=df.copy()
#syntax
df1.fillna('Empty',inplace=True)
display(df1.isnull().sum())
display(df1.head(10))
```

```python
#fill NaN value with 23 for 'Age' column
df1=df.copy()
#syntax
df1['Age'].fillna(23,inplace=True)
display(df1.isnull().sum())
```

```python
#fill NaN value in specific cell(2,'MetColor')
df1=df.copy()
#syntax
df1.at[2,'MetColor']=10
df1.head()
```

```python
#fill with a measure of central tendency in 'Age' column - mean, median, min, max, standard deviation
df1=df.copy()
display(df1.tail())
#syntax
df1['Age'].fillna(df1['Age'].mean(),inplace=True)
df1['FuelType'].fillna(df1['FuelType'].value_counts().index[0],inplace=True)
display(df1.isnull().sum())
display(df1.tail())
```

```python
#fill with value from the next row - bfill
df1=df.copy()
display(df1.head(10))
#syntax
df1.fillna(method='bfill',inplace=True)
display(df1.head(10))
```

```python
#fill with value from the next row for specific column (MetColor)
df1=df.copy()
display(df1.head(10))
#syntax
df1['MetColor'].fillna(method='bfill',inplace=True)
display(df1.head(10))
```

```python
#fill with value from the previous row - ffill
df1=df.copy()
display(df1.head(10))
#syntax
df1.fillna(method='ffill',inplace=True)
display(df1.head(10))
```

```python
#fill with value from the previous row for specific column (MetColor)
df1=df.copy()
display(df1.head(10))
#syntax
df1['MetColor'].fillna(method='ffill',inplace=True)
display(df1.head(10))
```

```python
#fill with value using interpolation method - linear, quadratic, polynomial (order), pad
df1=df.copy()
display(df1.head(10))
#syntax
# df1['MetColor'].interpolate(method='nearest',inplace=True)
df1['MetColor'].interpolate(method='polynomial',order=1,inplace=True)
display(df1.head(10))
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df_imputed = imputer.fit_transform(df)
print(df.head())
print(pd.DataFrame(df_imputed).head())
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_imputed = imputer.fit_transform(pd.DataFrame(df['MetColor']))
print(df['MetColor'].head())
print(pd.DataFrame(df_imputed).head())
```
